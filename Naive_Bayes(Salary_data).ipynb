```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

Insight :

-  Creating a function for ploting a data frame againts feature to see the linearity.
-  We use probplot here to to calculate best fit line and plot according to that.

```python
df.head(1)
```

```python
plot_data(df,'age')
```

```python
plot_data(df,'educationno')
```

```python
plot_data(df,'capitalloss')
```

```python
plot_data(df,'capitalgain')
```

```python
plot_data(df,'hoursperweek')
```

Insight :

-  Ploting all the features against data frame to see the best fit line.

```python
from sklearn.model_selection import train_test_split
```

Insight :

-  With the help of train_test split we divided our dataframe into two parts (train and test)
-  we divided our data into train and test data into 70-30 size means our train data size is 70% of the whole data and test data size is 30% of the whole data
-  we divided our data in two data set because we perform all EDA, visualization, taking some insights from train data, and taken sll insights will apply on test data to predict and to check how good our model is work

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
encoder = ce.OneHotEncoder(cols=['workclass','education','maritalstatus','occupation','relationship','race','sex','native'])
```

```python
x_train = encoder.fit_transform(x_train)
```

```python
x_test = encoder.transform(x_test)
```

Insights :

-  .catogory encoders is use to convert the catogorical object data into numeric format (0 and 1).

Robust Scaler

```python
cols = x_train.columns
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler  = RobustScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train = pd.DataFrame(x_train,columns=[cols])
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NaiveBayes_model = GaussianNB()
```

```python
NaiveBayes_model.fit(x_train,y_train)
```

```python
NaiveBayes_model.score(x_test,y_test)
```

```python
y_pred = NaiveBayes_model.predict(x_test)
```
